```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# **Configuraci√≥n del navegador**
options = Options()
options.add_argument("--disable-blink-features=AutomationControlled")  # Evitar detecci√≥n como bot
options.add_argument("--log-level=3")  # Reducir logs en consola
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# **Inicializar WebDriver**
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# **Abrir la URL de FIBA**
url = "https://www.fiba.basketball/en/events/basketball-champions-league-americas-24-25"
driver.get(url)

# **Esperar unos segundos para que la p√°gina cargue completamente**
time.sleep(5)

print("‚úÖ P√°gina abierta correctamente. Realiza el clic manualmente.")

# **La sesi√≥n del navegador sigue abierta para continuar con otras acciones**
```

```python
# **Paso 3: Hacer clic en el bot√≥n de "Teams"**
try:
    # **Buscar el bot√≥n usando el XPath proporcionado**
    teams_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="themeWrapper"]/div[2]/header/div[1]/div[2]/nav/div[2]/ul/li[4]'))
    )

    # **Hacer scroll hasta el bot√≥n por si est√° oculto**
    driver.execute_script("arguments[0].scrollIntoView();", teams_button)

    # **Intentar hacer clic en el bot√≥n**
    teams_button.click()

    print("‚úÖ Se hizo clic en el bot√≥n 'Teams' correctamente.")

    # **Esperar a que la p√°gina de equipos cargue**
    time.sleep(3)

except Exception as e:
    print(f"‚ùå No se pudo hacer clic en el bot√≥n 'Teams'. Error: {e}")
```

```python
# **Definir la ruta donde se guardar√° el archivo CSV**
ruta_guardado = r"C:\Users\anavarro\OneDrive - INDEPORTES ANTIOQUIA\AI\2024\METODOLOGO\PYTHON\MODULO 11 Proyeco FM\M_11_PROYECTO PAISAS\Data\df_info_teams.csv"

# **PASO 4: Esperar que el contenedor de equipos est√© disponible**
try:
    equipos_contenedor = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "_1njabzg0"))  # Grid de equipos
    )
    print("‚úÖ Contenedor de equipos detectado.")

except Exception as e:
    print(f"‚ùå No se encontr√≥ la secci√≥n de equipos. Error: {e}")

# **Lista para almacenar los datos extra√≠dos**
teams_data = []

# **Buscar todos los elementos de los equipos dentro del contenedor**
equipos = driver.find_elements(By.XPATH, '//*[@id="themeWrapper"]/div[2]/div/div/div/div[3]/div[1]/div/div/a')

# Confirmamos cu√°ntos equipos se detectaron
print(f"‚úÖ Se encontraron {len(equipos)} equipos.")

# **Iterar sobre cada equipo para extraer la informaci√≥n**
for equipo in equipos:
    try:
        # **Extraer Nombre del Club**
        nombre_equipo = equipo.find_element(By.XPATH, './div[2]/h2/span').text.strip()

        # **Extraer URL del Logo**
        logo_element = equipo.find_element(By.XPATH, './div[2]/div/div/img')
        logo_url = logo_element.get_attribute("src")

        # **Extraer Texto Adicional**
        texto_extra = equipo.find_element(By.XPATH, './div[2]/span/div').text.strip()

        # **Agregar los datos a la lista**
        teams_data.append([nombre_equipo, logo_url, texto_extra])
        print(f"‚úÖ Extra√≠do: {nombre_equipo} - {logo_url} - {texto_extra}")

    except Exception as e:
        print(f"‚ö†Ô∏è Error al extraer datos de un equipo. Detalles: {e}")

# **Crear el DataFrame**
df_info_teams = pd.DataFrame(teams_data, columns=["Team", "Url", "Name"])

# **Mostrar en la consola**
print("\n‚úÖ DataFrame generado:")
print(df_info_teams)

# **Guardar en la ruta especificada**
df_info_teams.to_csv(ruta_guardado, index=False)
print(f"‚úÖ Archivo guardado en: {ruta_guardado}")
```

```python

# **PASO 1: Identificar los equipos en el contenedor `<div class="_1njabzg0">`**
try:
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "_1njabzg0"))
    )
    print("‚úÖ Lista de equipos cargada correctamente.")
except Exception as e:
    print(f"‚ùå Error al cargar la lista de equipos: {e}")
    driver.quit()

# **PASO 2: Obtener la lista de equipos**
equipos = driver.find_elements(By.XPATH, '//*[@id="themeWrapper"]/div[2]/div/div/div/div[4]/div[1]/div/div/a')

if len(equipos) > 0:
    print(f"‚úÖ Se encontraron {len(equipos)} equipos.")
else:
    print("‚ùå No se encontraron equipos. Verifica el XPATH.")
    driver.quit()

# **PASO 3: Iterar sobre cada equipo, hacer clic y luego hacer clic en "Roster"**
for index, equipo in enumerate(equipos):
    try:
        print(f"\n‚û°Ô∏è Procesando equipo {index + 1} de {len(equipos)}...")

        # **Hacer clic en el equipo**
        equipo.click()
        print("‚úÖ Click en equipo realizado.")

        # **Esperar que la p√°gina del equipo cargue**
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="radix-:r6d:-trigger-roster"]'))
        )
        print(f"‚úÖ P√°gina del equipo cargada.")

        # **Hacer clic en el bot√≥n "Roster"**
        roster_button = driver.find_element(By.XPATH, '//*[@id="radix-:r6d:-trigger-roster"]')
        roster_button.click()
        print("‚úÖ Click en 'Roster' realizado.")

        # **Esperar a que la secci√≥n del roster cargue completamente**
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "roster-container"))  # Ajustar seg√∫n el HTML del roster
        )
        print("‚úÖ Secci√≥n de 'Roster' cargada.")

        # **Volver a la p√°gina de equipos**
        driver.back()
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "_1njabzg0"))
        )
        print("üîÑ Regresando a la p√°gina de equipos.")

    except Exception as e:
        print(f"‚ö†Ô∏è Error al procesar el equipo {index + 1}: {e}")
        driver.back()
        continue

# **Mantener el navegador abierto para pr√≥ximos pasos**
print("\n‚úÖ Proceso finalizado. El navegador sigue abierto para continuar con otros pasos.")
```
